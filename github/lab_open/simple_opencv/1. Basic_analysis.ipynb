## パスを直したけど色がおかしい

```python
img = cv2.imread('img/Lenna.png')
plt.imshow(img)
#デフォルトではBGR（青、緑、赤）の順番にimportされるが描画がRGB（赤、緑、青）なので配色がおかしくなる
#https://note.nkmk.me/python-opencv-bgr-rgb-cvtcolor/
```

## 色をなおす

```python
img = cv2.imread('img/Lenna.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)#convert BGR to RGB
print(img_rgb.shape)#3次元
plt.imshow(img_rgb)
```

## グレースケールへの変更

```python
img = cv2.imread('img/Lenna.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#convert BGR to GRAY
print(img_gray.shape)#２次元
plt.imshow(img_gray)
```

## データの保存

```python
cv2.imwrite('output/lena_gray.jpg', img_gray)
cv2.imwrite('output/lena_rgb.jpg', img_rgb)##最後にbgrになおさないとおかしくなるので注意
```

## 画像縦連結

```python
im_v = cv2.vconcat([img,img_rgb])
plt.imshow(im_v)
```

## 画像横連結

```python
im_h = cv2.hconcat([img,img_rgb])
plt.imshow(im_h)
```

## 同一画像の繰り返しならべ
[2]np.tileの説明　https://deepage.net/features/numpy-tile.html

```python
im_v_np = np.tile(img_rgb, (2, 5,1))#縦方向，横方向，３次元方向は１
plt.imshow(im_v_np)
```

## RGBの一色以外を０に固定する

```python
img_R = cv2.imread('img/Lenna.png')
img_R[:, :, (0, 1)] = 0##BGRのR以外を０
img_R = cv2.cvtColor(img_R, cv2.COLOR_BGR2RGB)#表示用に色を変更する
plt.imshow(img_R)
```

```python
img_G = cv2.imread('img/Lenna.png')
img_G[:, :, (0, 2)]= 0##BGRのG以外を０
img_G = cv2.cvtColor(img_G, cv2.COLOR_BGR2RGB)#表示用に色を変更する

img_B = cv2.imread('img/Lenna.png')
img_B[:, :, (1, 2)]= 0##BGRのB以外を０
img_B = cv2.cvtColor(img_B, cv2.COLOR_BGR2RGB)#表示用に色を変更する
im_h_RGB = cv2.hconcat([img_R,img_G,img_B])
plt.imshow(im_h_RGB)
```
